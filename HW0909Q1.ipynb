```
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
def convert2float(row=None):
  new_row=[]
  for e in row:
    new_row.append(float(e))
  return new_row

with open('/iris.csv') as csvfile:
  readCSV = csv.reader(csvfile, delimiter=',')
  M1=[]
  M2=[]
  M3=[]
  for row in readCSV:
    if row!=[]:
      if row[-1]=="Iris-setosa":
        M1.append(convert2float(row[:-1]))
      elif row[-1]=="Iris-versicolor":
        M2.append(convert2float(row[:-1]))
      elif row[-1]=="Iris-virginica":
        M3.append(convert2float(row[:-1]))

```

```
def getdim(M=M1):
  row=len(M)
  col=len(M[0])
  return row, col
```

```
#dimensions
dim1=getdim(M1)
dim2=getdim(M2)
dim3=getdim(M3)
print("the dimension of M1 is "+str(dim1[0])+"*"+str(dim1[1]))
print("the dimension of M2 is "+str(dim2[0])+"*"+str(dim2[1]))
print("the dimension of M1 is "+str(dim3[0])+"*"+str(dim3[1]))
```

```
def col_stats(M=M1):
  c_num=len(M[0])
  #print(c_num)
  col_avg=[]
  col_min=[]
  for i in range(c_num):
    clist=[]
    for row in M:
      clist.append(row[i])
    #print(clist)
    col_avg.append(sum(clist)/len(clist))
    col_min.append(min(clist))
  
  return col_avg, col_min

col_stats(M=M1)
```

```
# column averages and column minimums
col_stats=[col_stats(M1),col_stats(M2),col_stats(M3)]
for i in [1,2,3]:
  for j in range(len(M1[0])):
    #print(j)
    print("Average of M"+str(i)+" "+"column"+str(j+1)+": "+str(col_stats[i-1][0][j]))
    print("Minimum of M"+str(i)+" "+"column"+str(j+1)+": "+str(col_stats[i-1][1][j]))



```
